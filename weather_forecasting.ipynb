```python
import pandas as pd
import numpy as np
```

```python
weather_data=pd.read_csv("weather_dataset_stage1 - Sheet 1 - weather_dataset.csv") #import the file
weather_data
```

```python
if "Date1" in weather_data.columns:
    weather_data.drop(["Date1"], axis=1, inplace=True)
else:
    print("Column 'Date1' does not exist in the dataframe.")
#delete column Date1 
```

```python
weather_data
```

```python
weather_data.drop_duplicates(inplace=True) #delete duplicates values
weather_data
```

```python
columns = weather_data.columns
for i in columns:
    weather_data[i] = weather_data[i].apply(lambda X:X.replace("'",""))
    weather_data
    #removing '' from the data
```

```python
weather_data["Date"] = pd.to_datetime(weather_data["Date"])
weather_data
#changing format for date
#error so first remove invalid date
```
